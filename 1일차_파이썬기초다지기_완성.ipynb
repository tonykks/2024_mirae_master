```python
# isinstance()함수를 이용하여 데이터 타입 확인
def get_datatype(data):
    instance= {int:'정수', float:'실수', str:'문자열', bool:'불린',
           list:'리스트', tuple:'튜플', dict:'딕셔너리', set:'집합' }
    for key, value in instance.items():
        if isinstance(data, key):
            print(f'{data}는 {value} 형입니다.')

for data in datas:
    get_datatype(data)
```

### 실습문제 : 도둑잡기 게임 만들기
경찰에 쫓기던 범인이 3개의 방 중에 한 곳에 숨어버렸습니다. 범인은 방문을 열 때마다 무작위로 3개의 방 중 하나의 방으로 이동합니다. 3개의 방 중 하나의 방을 선택했을 때 범인이 있다면 100점, 없을 때 -10점을 부여하고 범인을 잡으면 게임을 종료하는 프로그램을 만들어 보세요.

```python
import random
score = 0

while True:  # 무한반복
    # 1.무작위로 범인 방 배치
    room = random.randint(1, 3)

    # 2.사용자로부터 방 번호 입력 받기
    n = int(input('방 번호를 입력하세요(1,2,3): '))

    # 3.사용자가 입력한 방 번호와 범인이 있는 방 비교
    if n == room:               # 4.범인을 찾으면
        print('범인 체포!')     # 범인 체포“라는 메시지 출력
        score += 100            # 점수 증가
        break                   # 무한반복 빠져 나가기

    else:                        # 5.범인을 찾지 못하면
        print('범인이 없습니다.' )# “범인을 찾지 못했다“라는 메시지 출력
        score -= 10               # 점수 감소

print('게임 종료!')        # 게임 종료 메시지 출력
print(f'점수: {score} 점') # 최종 점수 출력
```

### 실습문제 : 암스트롱수 판별 함수
다음 조건을 참고하여 암스트롱수를 구하는 함수를 만들어 출력하는 프로그램을 만들어 보세요.
- **암스트롱수** : 세 자리의 정수 중에서 각 자리의 수를 세 제곱한 수의 합과 자신이 같은 수

```python
def is_armstringNumber(N):
    _3, _2, _1 = int(str(N)[0]), int(str(N)[1]),  int(str(N)[2])
    if N == (_3**3 + _2**3 + _1**3):
        return True
    else:
        return False


for i in range(100, 1000):
    if is_armstringNumber(i):
        print("[%i] is armstrong number" % i)
```

```python
menu = '''---선택메뉴--
1. 단어 등록
2. 단어 검색
3. 단어 삭제
4. 모든 단어 보기
0. 종료하기
------------------------------'''

# 빈 딕셔너리를 만든다. (반복문 안에 넣지 않는다!)
wordbook = {}

def addBook():
    w = input('새 단어:')
    m = input(f'{w}의 뜻:')
    wordbook[w] = m

def searchBook():
    w = input('검색할 단어:')
    if w in wordbook:
        print(f'{w} : {wordbook[w]}')
    else:
        print(f'미등록 단어')

def deleteBook():
    w = input('삭제할 단어:')
    if w in wordbook:
        del wordbook[w]
        print(f'{w} 단어 삭제됨!')
    else:
        print(f'미등록 단어')

def showAllBook():
    for w, m in wordbook.items():
        print(f'{w} : {m}')


while True:
    select = input(menu+'선택:')

    if select == '0':
        print('프로그램을 종료합니다.')
        break
    elif select == '1':          #1. 단어 등록
        addBook()
    elif select == '2':          #2. 단어 검색
        searchBook()
    elif select == '3':          #3. 단어 삭제
        deleteBook()
    elif select == '4':          #4. 모든 단어 보기
        showAllBook()
    else:
        print('잘못 입력하셨습니다.')

    print('-'*30)
```
